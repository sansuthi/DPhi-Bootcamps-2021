# **Task2 : Visualization of results using Matplotlib library**

* Display the data of top 5 east asian countries based on 'Generosity'
* Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries
* Build a plot that shows the variation of 'Ladder score' among the above 5 south east asian countries
* Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'

```python
# Importing Library
import matplotlib.pyplot as plt
```

```python
# Unique values in the column - 'Regional indicator'
df1['Regional indicator'].unique()
```

```python
# Top 5 East Asian countries based on 'Generosity'
countries_EA_top5 = df1[df1['Regional indicator'] == 'East Asia'].sort_values('Generosity', ascending=False)[:5]
countries_EA_top5
```

```python
# Line plot that shows the variation of 'Ladder score' of top 5 east asian countries 
plt.figure(figsize=(12, 5))
plt.plot(countries_EA_top5['Country name'], countries_EA_top5['Ladder score'], '--ro')
plt.title('Ladder Score of Top 5 East Asian Countries')
plt.xlabel('COUNTRY', fontsize=12)
plt.ylabel('LADDER SCORE', fontsize=12)
plt.show()
```

```python
# Top 5 South East Asian countries based on 'Generosity'
countries_SEA_top5 = df1[df1['Regional indicator'] == 'Southeast Asia'].sort_values('Generosity', ascending=False)[:5]
countries_SEA_top5
```

```python
# Line plot that shows the variation of Ladder score' of top 5 south east asian countries
plt.figure(figsize=(12, 5))
plt.plot(countries_SEA_top5['Country name'], countries_SEA_top5['Ladder score'], '-.go')
plt.title('Ladder Score of Top 5 South East Asian Countries')
plt.xlabel('COUNTRY', fontsize=12)
plt.ylabel('LADDER SCORE', fontsize=12)
plt.show()
```

```python
# Creating the dataframe object - df_2021 with the following countries: 
# 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'.
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(countries)]
df_2021
```

```python
# scatter plot to show the relation the countries vs their 'Logged GDP per capita'
plt.figure(figsize=(10, 5))
plt.scatter(df_2021['Country name'], df_2021['Logged GDP per capita'])
plt.title('Logged GDP per capita')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Logged GDP per capita', fontsize=12)
plt.show()
```
